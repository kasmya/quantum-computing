```python
from qiskit import QuantumCircuit
from qiskit_aer.primitives import Sampler

# Create a circuit with 2 qubits (1 input, 1 output)
qc = QuantumCircuit(2)

# Step 1: Initialize output qubit to |1>
qc.x(1)

# Step 2: Apply Hadamard to both qubits
qc.h([0, 1])

# --- Oracle Example ---
# Constant function (does nothing)
# Balanced function (uncomment below)
# qc.cx(0, 1)

# Step 3: Apply Hadamard to input qubit
qc.h(0)

# Measure input qubit
qc.measure_all()

# Run with Sampler
sampler = Sampler()
result = sampler.run(qc, shots=1000).result()

print("Measurement Results:", result)
print(qc)
```
