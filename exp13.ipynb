```python
from qiskit import QuantumCircuit
from qiskit_aer.primitives import Sampler

# Build the circuit
qc = QuantumCircuit(2, 1)
qc.x(1)
qc.h([0, 1])
qc.cx(0, 1)   # balanced function
qc.h(0)
qc.measure(0, 0)

# Run with Sampler
sampler = Sampler()
result = sampler.run(qc, shots=1000).result()

print(result)
```
